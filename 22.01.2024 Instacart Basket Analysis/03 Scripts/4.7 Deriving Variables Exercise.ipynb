# 03. Create Function with Conditions

```python
# Define a function
def price_label(row):
    
    if row['prices']<= 5:
        return 'low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'mid-range product'
    elif row['prices'] > 15:
        return 'high-range product'
    else: return np.nan
```

```python
# Apply function to df
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check values in new column
df['price_range'].value_counts(dropna = False)
```

```python
# Check most expensive product in subset
df['prices'].max()
```

# 04. Create If-Statements with the loc() Function

```python
# Create Conditions
df.loc[df['prices'] > 15, 'price_range_loc'] = 'high-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

# 05. Create If-Statements with For-Loops

```python
# On which day do most orders take place? Check Frequency
# 0 = Saturday
df_merge['order_day_of_week'].value_counts(dropna = False)
```

```python
# Create 'busiest day' list
result = []

for value in df_merge['order_day_of_week']:
    if value == 0:
        result.append('busiest day')
    elif value == 4:
        result.append('least busy')
    else: result.append('regularly busy')
```

```python
# Check list
result
```

```python
# Combine list with df_merge within new column
df_merge['busiest_day'] = result
```

```python
# Check Frequency
df_merge['busiest_day'].value_counts(dropna = False)
```

```python
df_merge.head()
```

# 06. Task 4.7.2 Create 'busiest days' column

```python
# Create 'busiest days' list
result_2 = []

for value in df_merge['order_day_of_week']:
    if value in[0,1]:
        result_2.append('busiest days')
    elif value in[3,4]:
        result_2.append('least busy')
    else: result_2.append('regularly busy')
```

```python
# Combine list with df_merge within new column
df_merge['busiest_days'] = result_2
```

```python
# Check Frequency
df_merge['busiest_days'].value_counts(dropna = False)
```

```python
# Check order_day_of_week is 1 is labelled 'busiest_days'
df_merge.loc[df_merge['order_day_of_week'] == 1].head(10)
```

# 07. Task 4.7.4 Create 'busiest_period_of_day' column

```python
# On which hour of day do most orders take place? Check Frequency
df_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create 'busiest period of day' list (Group by 8 into 3 categories)
result_3 = []

for value in df_merge['order_hour_of_day']:
    if value in[10,11,14,15,13,12,16,9]:
        result_3.append('most orders')
    elif value in[17,8,18,19,20,7,21,22]:
        result_3.append('average orders')
    else: result_3.append('fewest orders')
```

```python
# Check list result_3
result_3
```

```python
# Combine list with df_merge within new column
df_merge['busiest_period_of_day'] = result_3
```

```python
# Check Frequency
df_merge['busiest_period_of_day'].value_counts(dropna = False)
```
